```python
plt.figure(figsize=(20.0,10.0))
plt.title("aggregate")
#plt의 x 축
plt.xlabel('date',fontsize='20')
# plt의 y 축
plt.ylabel('incidence',fontsize='20')
c = []
r = []
n = []
d = []
t = []
for row in range(0,101):
        t.append(data['test'][row+1]-data['test'][row])
        c.append(data['confirmed'][row+1]-data['confirmed'][row])
        r.append(data['released'][row+1]-data['released'][row])
        n.append(data['negative'][row+1]-data['negative'][row])
        d.append(data['deceased'][row+1]-data['deceased'][row])

plt.plot(data['date'][:101],t,label='test')
plt.plot(data['date'][:101],n,label='negative')
plt.plot(data['date'][:101],c,label='confirmed')
plt.plot(data['date'][:101],r,label='released')
plt.plot(data['date'][:101],d,label='deceased')

plt.legend(loc='upper right',fontsize='x-large')
plt.show()
```
