````python
# 주어진 target의 범위
bins = [0.2, 1.7, 2.5, 5]
labels = ["level_1", "level_2", "level_3"]
````

````python
def make_onehot(df, target, bins, labels):
    data = pd.cut(target['Target'],bins=bins,labels=labels)
    data.dropna(inplace=True)
    tpp = pd.get_dummies(data)
    tpp = df.join(tpp,how='right')
    return tpp
````

````python
data = make_onehot(df, target = df_target, bins= bins, labels=labels)
````

# 문제 3

````python
from sklearn.datasets import make_blobs
import random
import numpy as np
````

sklearn 의 make_blobs를 이용하여 만든 데이터 X가 있다. 해당 X를 시각화 하면 다음과 같다.

````python
''""" 
make_blobs 함수는 등방성 가우시안 정규분포를 이용해 가상 데이터를 생성한다. 이 때 등방성이라는 말은 모든 방향으로 같은 성질을 가진다는 뜻이다. 다음 데이터 생성 코드의 결과를 보면 make_classification 함수로 만든 가상데이터와 모양이 다른 것을 확인 할 수 있다. make_blobs는 보통 클러스링 용 가상데이터를 생성하는데 사용한다. make_blobs 함수의 인수와 반환값은 다음과 같다.

인수:

n_samples : 표본 데이터의 수, 디폴트 100
n_features : 독립 변수의 수, 디폴트 20
centers : 생성할 클러스터의 수 혹은 중심, [n_centers, n_features] 크기의 배열. 디폴트 3
random_state : 난수 발생 시드
cluster_std: 클러스터의 표준 편차, 디폴트 1.0
X : n_samples x n_features 크기의 배열
독립 변수
y : n_samples 크기의 배열
종속 변수

numpy
random.randn(array_size) : random variable을 array_size크기의 array에 채우는 함수
"""
# plt figure 크기 지정 [row,col]
plt.rcParams['figure.figsize'] = [6, 4]
X, y = make_blobs(n_samples=300, n_features=2, centers=3, random_state=1, cluster_std = 0.001)
X2, _ = make_blobs(n_samples=300, n_features=2, centers=1, random_state=1, cluster_std = 1)
X = np.vstack([X, X2])
stdev = 0.45
plt.scatter(X[:, 0]+np.random.randn(len(X[:, 0]))*stdev, X[:, 1]+np.random.randn(len(X[:, 1]))*stdev, marker='o', s=100, edgecolor="k", linewidth=2)
plt.show()
````
